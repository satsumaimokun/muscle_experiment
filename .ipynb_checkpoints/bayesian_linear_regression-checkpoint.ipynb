```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import pymc3 as pm
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('muscle_data.csv')
```

```python
df.head()
```

```python
x = df.drop(["ABS","nagano","fujino","maeda","yamasaki"],axis=1)
```

```python
#独立変数の標準化
sscaler = preprocessing.StandardScaler()
x_std = sscaler.fit_transform(df[["HR","percent"]]) 
x["HR"] = x_std[:,0]
x["percent"] =x_std[:,1]
```

```python
H = x["HR"].tolist()
T = df["ABS"].tolist() #従属変数: abs(指示された筋力(%)-被験者が出した筋力(%))
P = x["percent"].tolist()
x.name = x.name.map(str.strip)
name_unique = x.name.unique()
names = len(name_unique)
name_lookup = dict(zip(name_unique,range(len(name_unique))))
```

```python
name = x["name_code"] = x.name.replace(name_lookup).values
```

```python
# モデルの作成
with pm.Model() as model:
    #事前分布
    sigma_T = pm.Uniform('sigma_T', lower=0, upper=100)
    a = pm.Uniform("a", lower=-100, upper=100)
    c = pm.Uniform('c', lower=-100, upper=100)
    inter = pm.Uniform('inter', lower=-100, upper=100)    
    
    #パイパーパラメタ
    sigma_b = pm.Uniform("sigma_b", lower=0, upper=100)
    
    #個人差項
    b = pm.Normal("b", mu=0, sd=sigma_b, shape=names)
    
    mu_T = a*H + b[name] + c*P + inter #a*(心拍数) + (個人差項) + c*(筋力パーセント) + 切片
    
    T_obs = pm.Normal('T_obs', mu=mu_T, sd=sigma_T, observed=T)

#modelの図示    
pm.model_to_graphviz(model)
```

```python
#サンプリング
with model:
    trace=pm.sample(10000, step=pm.NUTS(), tune=1000, chains=4, random_seed=0)
```

```python
#waicの計算
waic = pm.waic(trace,model)
waic.WAIC
```

```python
#パラメタの事後分布の図示
with model:
    pm.traceplot(trace)
    plt.show()
```

```python
#事後分布の要約統計量
with model:
    summary_trace = pm.summary(trace)
    print(summary_trace)   
```

```python
#心拍数の係数が正である確率
post_a = trace["a"]
print((post_a>0).sum()/len(post_a))
```

```python
#予測分布からのサンプリング
ppc = pm.sample_posterior_predictive(trace, samples=5000, model=model)
```

```python
#1つめのデータ点に対応するヒストグラム
plt.hist(ppc["T_obs"][:,0],bins=100)
plt.show()
```

```python
#ヒスグラムの中央値と四分位範囲
median = np.percentile(ppc["T_obs"], 50, axis=0)
err_list = np.percentile(ppc["T_obs"], 75, axis=0) - np.percentile(ppc["T_obs"], 25, axis=0)
```

```python
#横軸に観測値、縦軸に予測値を取ったグラフの図示
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(1,1,1)

#エラーバーとして四分位範囲を描画
for obs, pred, err in zip(T, median, err_list):
    ax.errorbar(obs, pred, err, fmt='ro', capsize=4, ecolor='lightskyblue', mfc="cornflowerblue",  mec="cornflowerblue")

#理想的な直線(y=x)の描画    
l = np.linspace(0, 40, 4000)
ax.plot(l,l)

ax.set_xlim([-5,45])
ax.set_xlim([-5,45])

plt.show()
```

```python
#観測値と予測値の相関係数
np.corrcoef(T,median)[1,0]
```
